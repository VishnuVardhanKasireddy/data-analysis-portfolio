# Walmart sales Dataset

## 1. Problem Statement
Analyze Walmart sales data to identify trends, seasonal patterns, and factors affecting weekly sales across stores and departments.

## 2. DataSets
- train.csv → sales data
- features.csv → external factors (temperature, fuel, CPI, etc.)
- stores.csv → store metadata

## 3. Key Challengs
1. Missing Values
2. Duplicated Values
3. Time-Series Data
4. Multiple Tables of Relational Data
