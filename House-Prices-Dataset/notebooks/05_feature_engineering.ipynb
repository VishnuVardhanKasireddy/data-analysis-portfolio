# Feature Engineering

In this notebook, new features are created based on insights obtained during exploratory data analysis.  
Feature engineering helps improve the predictive power of models by transforming raw data into more meaningful variables.

```python
import numpy as np
import pandas as pd

df=pd.read_csv('C:/Dev/Projects/EDA-Portfolio/House-Prices-Dataset/Data/processed/cleaned_train.csv')
df.head()
```

Total living space is a strong indicator of house value.  
Combining basement and floor areas creates a more comprehensive measure of property size.

```python
df['TotalSF']=df['TotalBsmtSF']+df['1stFlrSF']+df['2ndFlrSF']
df['TotalBathrooms']=df['FullBath']+(0.5*df['HalfBath'])+df['BsmtFullBath']+(0.5*df['BsmtHalfBath'])
df['HouseAge']=2026-df['YearBuilt']
df['RemodAge']=2026-df['YearRemodAdd']
df['TotalPorchSF']=df['OpenPorchSF']+df['EnclosedPorch']+df['3SsnPorch']+df['ScreenPorch']+df['WoodDeckSF']
df['TotalRooms']=df['TotRmsAbvGrd']+df['TotalBathrooms']
```

```python
cols_to_drop=['Id','1stFlrSF','2ndFlrSF','TotalBsmtSF','FullBath','HalfBath','BsmtFullBath','BsmtHalfBath','YearBuilt','YearRemodAdd','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','WoodDeckSF','TotRmsAbvGrd']
df=df.drop(columns=cols_to_drop)
df.shape
```

## Feature Engineering Summary

Several new features were created to better capture structural characteristics of houses.

Key engineered features include:
- **TotalSF** – total square footage combining basement and floors
- **HouseAge** – age of the property
- **RemodelAge** – years since last remodeling
- **TotalBathrooms** – combined bathroom count
- **TotalPorchSF** – total porch area

These engineered variables provide richer representations of house characteristics and are expected to improve predictive model performance.
