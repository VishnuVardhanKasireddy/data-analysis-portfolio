# Model Development

In this notebook, machine learning models are trained to predict house prices using the engineered dataset.

Multiple regression models will be evaluated and compared to determine the most effective predictive approach.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv('C:/Dev/Projects/EDA-Portfolio/House-Prices-Dataset/Data/processed/engineered_train.csv')
df.head()
```

```python
x=df.drop(columns=['SalePrice'])
y=df['SalePrice']
```

```python
x=pd.get_dummies(x,drop_first=True)
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred_lr=lr.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_test,pred_lr))
print('Linear Regression RMSE:',rmse)
```

```python
ridge=Ridge()
ridge.fit(x_train,y_train)
pred_ridge=ridge.predict(x_test)
rmse_ridge=np.sqrt(mean_squared_error(y_test,pred_ridge))
print('Ridge Regression RMSE:',rmse_ridge)
```

```python
lasso=Lasso()
lasso.fit(x_train,y_train)
pred_lasso=lasso.predict(x_test)
rmse_lasso=np.sqrt(mean_squared_error(y_test,pred_lasso))
print('Lasso Regression RMSE:',rmse_lasso)
```

```python
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
pred_rf=rf.predict(x_test)
rmse_rf=np.sqrt(mean_squared_error(y_test,pred_rf))
print('Random Forest RMSE:',rmse_rf)
```

```python
results=pd.DataFrame({
    'Model':['Linear Regression','Ridge Regression','Lasso Regression','Random Forest'],
    'RMSE':[rmse,rmse_ridge,rmse_lasso,rmse_rf]
})
results.sort_values(by='RMSE')
```

```python
import joblib
joblib.dump(ridge,'c:/Dev/Projects/EDA-Portfolio/House-Prices-Dataset/Models/best_model.pkl')
joblib.dump((x_test,y_test),'c:/Dev/Projects/EDA-Portfolio/House-Prices-Dataset/Models/test_data.pkl')
```

```python
results.to_csv('c:/Dev/Projects/EDA-Portfolio/House-Prices-Dataset/Models/model_comparison.csv',index=False)
```

## Modeling Summary

Several regression models were trained and evaluated using RMSE as the performance metric.

Random Forest achieved the lowest error, suggesting that non-linear relationships exist between housing features and sale prices.

Regularized linear models such as Ridge and Lasso also performed well, indicating that feature engineering successfully captured key predictive information.
