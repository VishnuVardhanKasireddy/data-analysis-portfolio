# **Feature Engineering**
 Feature engineering the cleaned data for optimising the modeling of the data

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Titanic-Survival-Dataset/Data/processed/cleaned_data.csv')
df.head()
```

```python
df['FamilySize']=df['SibSp']+df['Parch']+1
df['IsAlone']=np.where(df['FamilySize']==1,1,0)
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df.head()
```

```python
df['Title'] = df['Title'].map({'Mr':0, 'Miss':1, 'Mrs':2, 'Master':3, 'Rare':4})
```

```python
df['AgeBand'] = pd.cut(df['Age'], bins=[0,12,20,40,60,80], labels=[0,1,2,3,4])
df=df.drop(['Name'], axis=1)
```

```python
df.head()
```

```python
df.to_csv('Titanic-Survival-Dataset/Data/processed/feature_engineered_data.csv', index=False)
```
