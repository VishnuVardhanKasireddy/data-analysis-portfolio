# **Exploratory Data Analysis**
- Exploring statiscal overview of the data
- univariate analysis
- bivariate analysis
- outlier analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic-Survival-Dataset/Data/processed/feature_engineered_data.csv')
df.head()
```

```python
df.describe()
```

## **Univariate Analysis**

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Survived' , data=df)
plt.title('Survival Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Fare'],kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
df['Fare']=np.log1p(df['Fare'])
plt.figure(figsize=(10,6))
sns.histplot(df['Fare'],kde=True)
plt.title('Log-Transformed Fare Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x=df['Parch'],kde=True)
plt.title('Parents/Children Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Sex', data=df)
plt.title('Count by Sex')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Pclass', data=df)
plt.title('Count by Passenger Class')
plt.show()
```

## **Bivariate Analysis**

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age by Survival')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Sex',y='Survived',data=df)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

## **Multivariate Analysis**

```python
plt.figure(figsize=(10,6))
sns.catplot(x='Pclass', y='Survived', hue='Sex',kind='bar', data=df)
plt.title('Survival Rate by Passenger Class ')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.catplot(x='Pclass', y='Survived', hue='IsAlone', kind='bar',data=df)
plt.title('Survival Rate by Alone Status and Sex')
plt.show()
```

```python
corr=df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Title',y='Survived',data=df)
plt.title('Survival Rate by Title')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.pairplot(df, hue='Survived')
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Fare vs Age by Survival')
plt.show()
```

# **Summary**
The analysis highlights that survival on the Titanic was not random but influenced by:

Gender

- Socio-economic status (Pclass, Fare)

- Age

- Family structure

##### Survival was strongly influenced by a combination of social hierarchy and evacuation priority, where women, children, and higher-class passengers had a clear advantage.
