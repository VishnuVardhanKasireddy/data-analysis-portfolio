# Modelling the Data
The goal is to build a classification model to predict passenger survival based on available features

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_auc_score,roc_curve,f1_score,precision_score,recall_score
```

```python
df=pd.read_csv('Titanic-Survival-Dataset/Data/processed/feature_engineered_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Title'])
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
print('x shape:',x.shape)
print('y shape:',y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.fit_transform(x_test)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr_pred=lr.predict(x_test)
print(f'Accuracy of logistic regression: %.2f' %( accuracy_score(y_test,lr_pred)*100))
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)
print(f'Accuracy of KNN: %.2f' %( accuracy_score(y_test,knn_pred)*100))
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt_pred=dt.predict(x_test)
print(f'Accuracy of Decision Tree: %.2f' %( accuracy_score(y_test,dt_pred)*100))
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
print(f'Accuracy of Random Forest: %.2f' %( accuracy_score(y_test,rf_pred)*100))    
```

```python
results=pd.DataFrame({
    'Model':['Logistic Regression','KNN','Decision Tree','Random Forest'],
    'Accuracy':[accuracy_score(y_test,lr_pred)*100,accuracy_score(y_test,knn_pred)*100,accuracy_score(y_test,dt_pred)*100,accuracy_score(y_test,rf_pred)*100]
}).sort_values(by='Accuracy',ascending=False)
```

```python
import joblib
joblib.dump(lr,'Titanic-Survival-Dataset/models/logistic_regression_model.pkl')
```
