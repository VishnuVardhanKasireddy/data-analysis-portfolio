# Titanic Survival Prediction
---

## 1. Problem Statement
    The Titanic disaster of 1912 is one of the most well-known maritime tragedies in history. During the incident, not all passengers were able to survive due to limited lifeboat capacity and evacuation constraints.

The goal of this project is to analyze passenger data from the Titanic and build a machine learning model capable of predicting whether a passenger would survive the disaster based on available features such as demographic information, ticket class, fare, and family relationships.

By performing exploratory data analysis and feature engineering, this study aims to identify the most important factors that influenced passenger survival. The project then applies classification algorithms to develop predictive models and evaluate their performance in determining survival outcomes.

The results of this analysis can help uncover patterns in survival probability and demonstrate how data science techniques can be used to analyze historical events and make predictive insights from structured datasets.

---

## Project Objectives
The main objectives of this project are:

- To analyze passenger characteristics and identify factors that influenced survival.

- To perform exploratory data analysis to understand relationships between variables.

- To clean and preprocess the dataset for machine learning.

- To engineer meaningful features that improve predictive performance.

- To train and compare classification models for predicting passenger survival.

## Dataset Description
### **Dataset Size**
The dataset is divided into two files on Kaggle:

#### **Training Dataset**

- 891 passengers

- Includes the target variable (Survived)

#### **Test Dataset**

- 418 passengers

- Used for prediction tasks

- Does not contain the survival column

 **Total passengers across both datasets:** 1309

---

## Type of Problem
This problem is a **binary classification** task where the objective is to predict whether a passenger survived the Titanic disaster based on demographic and travel-related features.
